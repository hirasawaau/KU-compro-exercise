```python
class MyMath:
    pi:float

    def __init__(self) -> None:
        self.pi = MyMath.FindPi()
    def is_even(self,r):
        return r%2 == 0
    def fac(self,r):
        return MyMath.findFact(r)

    def is_prime(self,r):
        for i in range(2,r):
            if r%i == 0:
                # print('This number is not a prime number.')
                return False
        # print('This number is a prime number.')
        return True

    def divide_by(self,n,k):
        return n%k == 0

    def ten_to_bi(self,r):
        base = bin(r)[2:]
        return base

    def ten_to_oct(self,n):
        base = oct(n)[2:]
        return base

    def ten_to_sixteen(self,n):
        base = hex(n)[2:]
        return base.upper()

    def int_to_roman(self,n):
        return MyMath.findRoman(n)
    


            
    @staticmethod
    def findFact(n:int):
        s = 1
        for i in range(n,0,-1):
            s = s * i
        return s
    
    @staticmethod
    def findRoman(n:int):
        RomanDict = {1:'I',5:'V',9:'IX',10:'X',40:'XL',50:'L',90:'XC',100:'C',400:'CD',500:'D',900:'CM',1000:'M'}
        romanItem = list(RomanDict.items())
        romanItem.reverse()
        st = ''
        for roman in romanItem:
            r = n // roman[0]
            if r > 0:
                st+=(roman[1])*r
                n-=(roman[0])*r
        
        return st

    @staticmethod
    def FindPi():
        listLoop = [i for i in range(2,101,2)]
        p = sum([1/((value) * (value+1) * (value+2)) if (index+1)%2 == 1 else -(1/((value) * (value+1) * (value+2))) for index,value in enumerate(listLoop)])
        return (p*4)+3
```

```python
my_math = MyMath()
num = int(input("Please Enter Number : "))
k = int(input("Divide by? : "))

if my_math.is_even(num):
    print('This number is even number.')
else:
    print('This number is not even number.')

if num <= 20:
    print(f'factorial is {my_math.fac(num):,.0f}.')
else:
    print('factorial TOO LONG')

if my_math.is_prime(num):
    print('This number is a prime number.')
else:
    print('This number is not a prime number.')

if(my_math.divide_by(num,k)):
    print(f'{num} is divisible by {k}')
else:
    print(f'{num} is not divisible by {k}')

print(f'{num} is {my_math.ten_to_bi(num)} in base 2.')
print(f'{num} is {my_math.ten_to_oct(num)} in base 8.')
print(f'{num} is {my_math.ten_to_sixteen(num)} in base 16.')
print(f'{num} is {my_math.int_to_roman(num)} in roman numeral system.')
print(f'PI is {my_math.pi:.20f}')
```
