5) เขียนโปรแกรมเพื่อหาจำนวนคนที่ยอดขายของอาหารสดรวมกับอาหารแช่แข็งผ่านช่องทาง Horeca(Hotel/Restaurant/Cafe) ไม่เกิน 1,000 ชิ้น

## 0) Prepare data

````python
rawDatas = open('wholesale.csv').read().strip().split('\n')
datas = [dict(zip(rawDatas[0].split(',') , rawData.split(','))) for rawData in rawDatas[1:]]
````

````python
datas
````

## 1) แปลงข้อมูลช่องทางการขายและภูมิภาค จากตัวเลขให้เป็นชื่อของแต่ละช่องทางการขายและภูมิภาคนั้น ๆ 

คอลัมน์ CHANNEL เป็นช่องทางการขาย ที่มีค่าเป็น Horeca (Hotel/Restaurant/Cafe) หรือ Retail (1,2)

คอลัมน์ REGION: เป็นภูมิภาค ได้แก่ Lisnon, Oporto หรือ Other (1,2,3)

````python
def changeData(data:dict):
    data['Channel'] = 'Horeca' if data['Channel'] == '1' else 'Retail'
    data['Region'] = 'Lisnon' if data['Region'] == '1' else 'Oporto' if data['Region'] == '2' else 'Other'
    return data
````

````python
import copy
translatedData = copy.deepcopy(list(map(changeData,datas)))
````

````python
translatedData
````

## 2) หายอดขายรวมจากทุกช่องทาง ในทุกภูมิภาค สำหรับสินค้าแต่ละประเภท



```
Fresh               5280131             
Milk                2550357             
Grocery             3498562             
Frozen              1351650             
Detergents_Paper    1267857             
Delicassen          670943 
```

````python
selllist = {}

for data in datas:
    for item in data.items():
        if item[0] == 'Channel' or item[0] == 'Region':
            continue
        try:
            selllist[item[0]]+=int(item[1])
        except:
            selllist.update({item[0]:int(item[1])})
        
````

````python
selllist
````

## 3) หายอดขายรวมของสินค้าทุกประเภท ในทุกช่องทาง สำหรับแต่ละภูมิภาค

```
Other 10677599
Lisnon 2386813
Oporto 1555088
```

````python
sellData = {}

for data in datas:
    summary = 0
    for value in data.values():
        try:
            summary+=int(value)
        except:
            continue
    try:
        sellData[data['Region']]+=summary
    except:
        sellData.update({data['Region']:summary})
````

````python
sellData
````

## 5) จำนวนคนที่ยอดขายของอาหารสดรวมกับอาหารแช่แข็งผ่านช่องทาง Horeca(Hotel/Restaurant/Cafe) ไม่เกิน 1,000 ชิ้น

```
10
```

````python
def filterData(data:dict):
    if data['Channel'] != 'Horeca':
        return False
    
    return int(data['Fresh']) + int(data['Frozen']) <= 1000
    
        
````

````python
filteredData = list(filter(filterData,datas))
````

````python
len(filteredData)
````

## 6) เขียนโปรแกรมเพื่อหาภูมิภาคที่มีการบริโภค นม มากที่สุด และในภูมิภาคนั้นมีการบริโภค นม มากกว่าอาหารสดทั้งหมดกี่คน

```
ภูมิภาคที่มีการบริโภค นม มากที่สุด
Other: 1888759

ภูมิภาคนั้นมีการบริโภค นม มากกว่าอาหารสดทั้งหมดกี่คน
107
```

````python
milkData = {}

for data in datas:
    try:
        milkData[data['Region']]+=int(data['Milk'])
    except:
        milkData.update({data['Region']:int(data['Milk'])})
````

````python
maxMilk = max(milkData.items(),key=lambda data:data[1])
maxMilk
````

````python
def FilterMilkMoreFresh(region:str):
    def _FilterMilkMoreFresh(data:dict):
        if data['Region'] != region:
            return False
        return int(data['Fresh']) < int(data['Milk'])
    return _FilterMilkMoreFresh
````

````python
filteredMilkMoreFreshData = list(filter(FilterMilkMoreFresh(maxMilk[0]) , datas))
````
