```python
def readFile(fileName):
    rawData = open(fileName).read().strip().split('\n')
    return [dict(zip(rawData[0].split(','),data.split(','))) for data in rawData[1:]]

def Menu1(datas:list):
    total = 0
    for data in datas:
        total+=int(data['Total Cases'])

    return total

def Menu2(datas:list):
    maxData = max(datas,key=lambda data:int(data['Deaths']))
    return maxData['State/UTs']

def Menu3(datas:list):
    def findMax(data:dict):
        percent = (int(data['Deaths']) / int(data['Total Cases'])) * 100
        return percent
    maxData = max(datas,key=findMax)
    return maxData['State/UTs']

def Menu4(datas:list):
    def findMax(data:dict):
        percent = (int(data['Discharged']) / int(data['Total Cases'])) * 100
        return percent
    maxData = max(datas,key=findMax)
    return maxData['State/UTs']

def Menu5(datas:list):
    active = sum([int(data['Active']) for data in datas])
    total = sum([int(data['Total Cases']) for data in datas])
    return active/total * 100

def Menu6(datas:list):
    def keyFunction(data:dict):
        day_avg = sum(([int(data[f'day{i}']) for i in range(1,8)])) / 7
        return day_avg
    def mapFunction(data:dict):
        day_avg = sum(([int(data[f'day{i}']) for i in range(1,8)])) / 7
        return {'State/UTs':data['State/UTs'],'avg':day_avg}
    sortedCovid = sorted(datas,key=keyFunction,reverse=True)
    return list(map(mapFunction,sortedCovid[0:5]))

def Menu7(datas:list):
    def keyFunction(data:dict):
        for i in range(1,6):
            if int(data[f'day{i}']) > 100:
                if i == 5:
                    return True
            else:
                break
        for i in range(2,7):
            if int(data[f'day{i}']) > 100:
                if i == 6:
                    return True
            else:
                break
        for i in range(3,8):
            if int(data[f'day{i}']) > 100:
                if i == 7:
                    return True
            else:
                return False
    
    filteredRed = list(filter(keyFunction,datas))
    return sorted([data['State/UTs'] for data in filteredRed] , key= lambda A:ord(A[0]))

fileName = input('filename: ')
Menu = int(input('type: '))

if Menu == 1:
    print(Menu1(readFile(fileName)))
elif Menu ==2:
    print(Menu2(readFile(fileName)))
elif Menu ==3:
    print(Menu3(readFile(fileName)))
elif Menu == 4:
    print(Menu4(readFile(fileName)))
elif Menu == 5:
    print(f'{Menu5(readFile(fileName)):.2f}%')
elif Menu == 6:
    datas = Menu6(readFile(fileName))
    for data in datas:
        if int(str(data['avg']).split('.')[1][0]) >= 5:
            data['avg']+=1
        print(f"{data['State/UTs']} {int(data['avg'])}")
elif Menu == 7:
    datas = Menu7(readFile(fileName))
    for data in datas:
        print(data)
```
