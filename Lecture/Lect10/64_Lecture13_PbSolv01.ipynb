````python
import numpy as np


x = [68.70, 31.53, 16.94, 9.95, 52.55, 29.65, 64.01, 69.52, 30.08, 21.77]
X = np.array(x)
print(X)

n = len(X)
mean = sum(X)/n
stdev = np.sqrt(sum((X-mean)**2) / (n-1))
print(f"Mean of the values is {mean:.2f}")
print(f"Standard deviation of the values is {stdev:.2f}")
````

### Pb02: Find the Top Most Bursting Hashtag

**Acknowledgement**: *Predicting Bursts and Popularity of Hashtags in Real-Time*, Kong et al., SIGIR'14 [click me](https://dl.acm.org/doi/pdf/10.1145/2600428.2609476)

#### We first find the time gap between the 1st and the last tweets.

````python
for i in range(len(tweets)):
  if i < 5:
    print(tweets[i]['created_at'])
  elif i > len(tweets)-5:
    print('>',tweets[i]['created_at'])
````

#### Define the helper function

````python
def calInterval(t1, t2):
  t1 = t1.split()[3].split(":")
  t2 = t2.split()[3].split(":")
  return (int(t2[0])-int(t1[0]))*60 + (int(t2[1])-int(t1[1]))*60 + int(t2[2])-int(t1[2])

def initCount(k=1800):
  return [0 for i in range(k)]

### test here
t1,t2 = 'Mon Sep 27 14:02:20 +0000 2021', 'Mon Sep 27 14:32:19 +0000 2021'
calInterval(t1, t2)
````

#### Plot the Bursting Hashtag 

````python
def extractHashtag_RetweetStatus2(tweets):
  my_hashtag = {}
  t1 = 'Mon Sep 27 14:02:20 +0000 2021'
  for tw in tweets:
    if tw.get('retweeted_status', None) != None:
      ht = ''
      for i in tw['retweeted_status']['entities']['hashtags']:
        ht += i['text'] + ","
        if my_hashtag.get(i['text'].upper(),None) != None:
          #my_hashtag[i['text'].upper()] += 1
          index = calInterval(t1, tw['created_at'])
          my_hashtag[i['text'].upper()][index] += 1
        else:
          #my_hashtag[i['text'].upper()] = 1
          my_hashtag[i['text'].upper()] = initCount(1800)
          index = calInterval(t1, tw['created_at'])
          my_hashtag[i['text'].upper()][index] = 1
  return my_hashtag

### main begins here
retweet_ht = extractHashtag_RetweetStatus2(tweets)
````
